```python
SELECT ROUND(COUNT(marital)/(select COUNT(marital) FROM 'Marketing Data.csv' )*100,2) AS divorced_percentage FROM 'Marketing Data1.csv' 
WHERE marital ='divorced' AND balance >2000;
```

```python
-- Query showing the percentage of divorced customers with 
-- a balance greater than 2000

SELECT
    ROUND((COUNT(CASE WHEN marital = 'divorced' AND balance > 2000 THEN 1 END) * 100/ COUNT(*)),2) AS Percentage
FROM
    'Marketing Data1.csv'
```

```python
-- Query showing the total count of each job in descending order

SELECT job,COUNT(*) AS total_job 
FROM 'Marketing Data1.csv'
GROUP BY job
ORDER BY total_job DESC
```

```python
-- QUery showing the educational level that gets the management job the most.

SELECT education, job,COUNT(job) AS Job_total
FROM 'Marketing Data1.csv'
WHERE job = 'management'
GROUP BY education, job
ORDER BY job_total DESC
LIMIT 1

```

```python
-- Query showing the average duration of customers in management 
-- positions calculated in years.

SELECT ROUND(AVG(duration/52),1) AS average_duration_inyears
FROM 'Marketing Data.csv'
WHERE job = 'management'
```

```python
-- Query showing the total number of customers that have housing, loan and are single.

SELECT housing,loan,marital, COUNT(marital) AS customer_count
FROM 'Marketing Data.csv' 
WHERE housing = 'yes' AND loan = 'yes' AND marital = 'single'
GROUP BY  housing,loan,marital
```

```python
SELECT housing,loan,marital, COUNT(marital) AS customer_count
FROM 'Marketing Data.csv' 
WHERE housing = 'yes' AND loan = 'yes' AND marital = 'single'
GROUP BY  housing,loan,marital 
```

```python
SELECT
	loan,marital,
    MIN(Age) AS MinAge,
    MAX(Age) AS MaxAge
FROM
    "Marketing Data.csv"
WHERE
    loan = 'yes' AND marital = 'single'
GROUP BY loan, marital;
```

```python
SELECT MIN(daily_decrease) AS highest_daily_decrease,
MAX(daily_increase) AS highest_daily_increase

FROM (SELECT _date,MIN(_close-_open) 
AS daily_decrease, MAX(_close-_open) 
AS daily_increase FROM 'SharePrice.csv'
GROUP BY _date) AS Price_changes
```

```python
SELECT
    MIN(DailyDecrease) AS highest_daily_decrease,
	MAX(DailyIncrease) AS highest_daily_increase
FROM (
    SELECT
        _date,
        MAX(_close - _open) AS DailyIncrease,
        MIN(_close - _open) AS DailyDecrease
    FROM 'SharePrice.csv'
    GROUP BY _date
) AS DailyChanges;
```

```python
SELECT first_name, last_name, department, salary,
DENSE_RANK() OVER(PARTITION BY department ORDER BY salary DESC) 
AS department_salary_rank
FROM "Employee-Table.csv"
ORDER BY department
```

```python
WITH salaryranks AS(SELECT first_name,last_name,
department,salary,
DENSE_RANK() OVER(PARTITION BY department ORDER BY salary DESC)AS department_salary_rank
FROM "Employee-Table.csv")
SELECt first_name,last_name,department,salary,department_salary_rank
FROM salaryranks
ORDER BY department
```
